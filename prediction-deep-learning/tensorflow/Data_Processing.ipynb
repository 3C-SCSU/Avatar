```python
import os

directory = r"C:\Users\student\Desktop\AV\brainwaves-csv\backward"
desired_rows = 1250
total_files_processed = 0

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        num_rows = len(df)

        if num_rows > desired_rows:
            truncated_rows = num_rows - desired_rows
            df = df.iloc[:desired_rows]
            df.to_csv(filepath, index=False)
            print(
                f"{filename}: Truncated {truncated_rows} rows. Total rows before truncation: {num_rows}"
            )
        else:
            print(f"{filename}: No truncation needed. Total rows: {num_rows}")

        total_files_processed += 1

print(f"Total files processed: {total_files_processed}")
```

```python
import os
import pandas as pd
import numpy as np

directory = r"C:\Users\student\Desktop\AV\brainwaves-csv\backward"
desired_rows = 1250
total_files_processed = 0

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        num_rows = len(df)

        if num_rows < desired_rows:
            padding_rows = desired_rows - num_rows
            padding_data = np.zeros((padding_rows, len(df.columns)))
            padding_df = pd.DataFrame(padding_data, columns=df.columns)
            df = pd.concat([df, padding_df], ignore_index=True)
            df.to_csv(filepath, index=False)
            print(
                f"{filename}: Padded with {padding_rows} rows of zeros. Total rows before padding: {num_rows}"
            )
        else:
            truncated_rows = num_rows - desired_rows
            df = df.iloc[:desired_rows]
            df.to_csv(filepath, index=False)
            print(
                f"{filename}: Truncated {truncated_rows} rows. Total rows before truncation: {num_rows}"
            )

        total_files_processed += 1

print(f"Total files processed: {total_files_processed}")
```
